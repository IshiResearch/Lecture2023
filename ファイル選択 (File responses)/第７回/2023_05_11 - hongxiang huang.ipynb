```python
img8=np.zeros((100,100),dtype = np.uint8)
for raw in range(100):
        img8[raw,:] = 99-raw
plt.imshow(img8,cmap = 'gray')
```

```python
img9 = np.zeros((100,100),dtype = np.uint8)
for col in range(100):
        img9[:,col] = 99-col    
plt.imshow(img9,cmap = 'gray')
```

```python
img10 = np.zeros((100,100),dtype = np.uint8)
for col in range(100):
        img10[:,col] = col
plt.imshow(img10,cmap = 'gray')
```

```python
img11 = np.zeros((100,100),dtype = np.uint8)
for row in range(100):
    for col in range(100):
        if row>col:
            img11[row,:col+1] = 99-row
        if col>row:
            img11[:row+1,col] = 99-col
        else:
            img11[row,col] = 99-row
    

plt.imshow(img11,cmap = 'gray')       
```

```python
img12 = np.zeros((255,255,3),dtype = np.uint8)
img12[0:255,0:(255//7),0] = 255
img12[0:255,(255//7):(255//7*2),1] = 255
img12[0:255,(255//7*2):(255//7*3),2] = 255
img12[0:255,(255//7*3):(255//7*4),0] = 255
img12[0:255,(255//7*3):(255//7*4),1] = 255
img12[0:255,(255//7*4):(255//7*5),0] = 255
img12[0:255,(255//7*4):(255//7*5),2] = 255
img12[0:255,(255//7*5):(255//7*6),1] = 255
img12[0:255,(255//7*5):(255//7*6),2] = 255
img12[0:255,(255//7*6):255,2] = 150
img12[0:255,(255//7*6):255,1] = 150
img12[0:255,(255//7*6):255,0] = 150
plt.imshow(img12)
```

```python
img13 = np.zeros((255,255,3),dtype = np.uint8)
img13[:,:]=255
for raw in range(255):
        img13[raw,0:(255//7),1] = 255-raw
        img13[raw,0:(255//7),2] = 255-raw
for raw in range(255):
        img13[raw,(255//7):(255//7*2),0] = 255-raw
        img13[raw,(255//7):(255//7*2),2] = 255-raw
for raw in range(255):
        img13[raw,(255//7*2):(255//7*3),0] = 255-raw
        img13[raw,(255//7*2):(255//7*3),1] = 255-raw
for raw in range(255):
        img13[raw,(255//7*3):(255//7*4),2] = 255-raw
for raw in range(255):
        img13[raw,(255//7*4):(255//7*5),1] = 255-raw
for raw in range(255):
        img13[raw,(255//7*5):(255//7*6),0] = 255-raw  
for raw in range(255):
        img13[raw,(255//7*6):255,0] = 255-raw//2
        img13[raw,(255//7*6):255,1] = 255-raw//2
        img13[raw,(255//7*6):255,2] = 255-raw//2 

plt.imshow(img13)        
```
