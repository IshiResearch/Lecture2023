```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 赤色画像
mat3d_red = np.zeros((50, 50, 3),np.uint8)
mat3d_red[:, :, 0] = 255
plt.imshow(mat3d_red)
```

```python
# 緑色画像
mat3d_green = np.zeros((50, 50, 3),np.uint8)
mat3d_green[:, :, 1] = 255
plt.imshow(mat3d_green)
```

```python
# 青色画像
mat3d_blue = np.zeros((50, 50, 3),np.uint8)
mat3d_blue[:, :, 2] = 255
plt.imshow(mat3d_blue)
```

```python
# シアン画像
mat3d_cyan = np.zeros((50, 50, 3),np.uint8)
mat3d_cyan[:, :, 1] = 255
mat3d_cyan[:, :, 2] = 255
plt.imshow(mat3d_cyan)
```

```python
# マゼンタ画像
mat3d_magenta = np.zeros((50, 50, 3),np.uint8)
mat3d_magenta[:, :, 0] = 255
mat3d_magenta[:, :, 2] = 255
plt.imshow(mat3d_magenta)
```

```python
# イエロー画像
mat3d_yellow = np.zeros((50, 50, 3),np.uint8)
mat3d_yellow[:, :, 0] = 255
mat3d_yellow[:, :, 1] = 255
plt.imshow(mat3d_yellow)
```

```python
# 画像を4分割し、シアン、マゼンタ、イエロー、灰色で表示
mat = np.zeros((100, 100, 3),np.uint8)

mat[:50, 50:, 2] = 255
mat[:50, 50:, 0] = 255

mat[:50, :50, 1] = 255
mat[:50, :50, 2] = 255

mat[50:, :50, 0] = 255
mat[50:, :50, 1] = 255

mat[50:, 50:] = 127
plt.imshow(mat)
```

```python
# 上から下に白色から黒色に変わる画像
img = np.zeros((256, 256), np.uint8)
number = 255

for i in range(256):
    img[i] = number
    number -= 1
    
plt.imshow(img, "gray")
```

```python
# 左から右に白色から黒色に変わる画像
img = np.zeros((256, 256), np.uint8)

for i in range(256):
    number = 255
    for j in range(256):
        img[i][j] = number
        number -= 1
        
plt.imshow(img, "gray")
```

```python
# 左から右に黒色から白色に変わる画像
img = np.zeros((256, 256), np.uint8)

for i in range(256):
    for j in range(256):
        img[i][j] = j
        
plt.imshow(img, "gray")
```

```python
# 左上から右下にかけて白色から黒色に変わる画像
img = np.zeros((256, 256), np.uint8)

for i in range(256):
    for j in range(256):
        img[i][j] = 255 - (i+j)//2
    
plt.imshow(img, "gray")
```

```python
# 左から右に黒色から赤色に変わる画像
img = np.zeros((256, 256, 3), np.uint8)

for i in range(256):
    img[:, i, 0] = i
        
plt.imshow(img)
```

```python
# 上から下に白色から青色に変わる画像
img = np.zeros((256, 256, 3), np.uint8)


for i in range(256):
    img[i, :] = 255 - i
    img[:,:, 2] = 255
        
plt.imshow(img)
```

```python
# 左から右に白色からマゼンタに変わる画像
img = np.zeros((256, 256, 3), np.uint8)

img[:,:] = 255

for i in range(256):
    img[:, i, 1] = 255 - i
    
        
plt.imshow(img)
```

```python
# 縦に７分割して、順に並ぶ画像
img = np.zeros((350, 350, 3), np.uint8)

img[:, :50, 0] = 255
img[:, 50:100, 1] = 255
img[:, 100:150, 2] = 255

img[:, 150:200, 0] = 255
img[:, 150:200, 1] = 255

img[:, 200:250, 0] = 255
img[:, 200:250, 2] = 255

img[:, 250:300, 2] = 255
img[:, 250:300, 1] = 255

img[:, 300:350] = 127

plt.imshow(img)
```

```python
# 11番目の問題の画像を上から下に、白色から徐々に色が変わる画像
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, :50, 0] = 255
    img[:, 50:100, 1] = 255
    img[:, 100:150, 2] = 255

    img[:, 150:200, 0] = 255
    img[:, 150:200, 1] = 255

    img[:, 200:250, 0] = 255
    img[:, 200:250, 2] = 255

    img[:, 250:300, 2] = 255
    img[:, 250:300, 1] = 255

    img[i:, 300:350] = 255 - (i//2)
plt.imshow(img)
```

```python
# 12番目の問題で作成した画像を色ごとに表示
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, :50, 0] = 255
img[:, 50:] = 0
    
plt.imshow(img)
```

```python
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, 50:100, 1] = 255
img[:, :50] = 0
img[:, 100:] = 0

plt.imshow(img)
```

```python
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, 100:150, 2] = 255
img[:, :100] = 0
img[:, 150:] = 0
plt.imshow(img)
```

```python
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, 150:200, 0] = 255
    img[:, 150:200, 1] = 255
img[:, :150] = 0
img[:, 200:] = 0
plt.imshow(img)
```

```python
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, 200:250, 0] = 255
    img[:, 200:250, 2] = 255
img[:, :200] = 0
img[:, 250:] = 0
plt.imshow(img)
```

```python
img = np.zeros((350, 350, 3), np.uint8)
for i in range(256):
    
    img[i: :] = 255 - i

    img[:, 250:300, 2] = 255
    img[:, 250:300, 1] = 255
img[:, :250] = 0
img[:, 300:] = 0
plt.imshow(img)
```
