## 4.上から下に白色から黒色に変化する画像

```python
mat9 = np.zeros((256,256),np.uint8)
for i in range(256):
    mat9[255-i] = i
plt.imshow(mat9, "gray")
```

## 5.左から右に白から黒に変化する画像

```python
mat10 = np.zeros((256,256),np.uint8)
for i in range(256):
    mat10[:,255-i] = i
plt.imshow(mat10, "gray")
```

## 6.左から右に黒から白に変化する画像

```python
mat11 = np.zeros((256,256),np.uint8)
for i in range(256):
    mat11[:,i] = i
plt.imshow(mat11, "gray")
```

## 7.左上から右下に白から黒に変化する画像

```python
mat12 = np.zeros((128,128),np.uint8)
for i in range(128):
    for j in range(128):
        mat12[i,j] = 255-(i+j)
plt.imshow(mat12, "gray")
```

## 8.左から右に黒から赤に変化する画像

## 9.上から下に白から青に変化する画像

```python
mat14 = np.zeros((256,256,3),np.uint8)
for i in range(256):
    mat14[:,:,2] = 255
    mat14[i,:,0] = (255-i)
    mat14[i,:,1] = (255-i)
plt.imshow(mat14, "gray")
```

## 11.縦に7分割して赤、緑、青、黄、マゼンタ、シアン、灰が順番に並ぶ画像

```python
mat16 = np.zeros((350,350,3),np.uint8)
mat16[:,0:50,0] = 255

mat16[:,50:100,1] = 255

mat16[:,100:150,2] = 255

mat16[:,150:200,0] = 255
mat16[:,150:200,1] = 255

mat16[:,200:250,0] = 255
mat16[:,200:250,2] = 255

mat16[:,250:300,1] = 255
mat16[:,250:300,2] = 255

mat16[:,300:350,0] = 100
mat16[:,300:350,1] = 100
mat16[:,300:350,2] = 100

plt.imshow(mat16, cmap = "gray")
```

## 12.上から下に白から徐々に色が変化する画像

```python
mat17 = np.zeros((255,350,3),np.uint8)
for i in range(255):
    mat17[:,0:50,0] = 255
    mat17[i,0:50,1] = (255-i)
    mat17[i,0:50,2] = (255-i)
    
    mat17[i,50:100,0] = (255-i)
    mat17[:,50:100,1] = 255
    mat17[i,50:100,2] = (255-i)
    
    mat17[i,100:150,0] = (255-i)
    mat17[i,100:150,1] = (255-i)
    mat17[:,100:150,2] = 255

    mat17[:,150:200,0] = 255
    mat17[:,150:200,1] = 255
    mat17[i,150:200,2] = (255-i)

    mat17[:,200:250,0] = 255
    mat17[i,200:250,1] = (255-i)
    mat17[:,200:250,2] = 255

    mat17[i,250:300,0] = (255-i)
    mat17[:,250:300,1] = 255
    mat17[:,250:300,2] = 255

    mat17[i,300:350,0] = 255-i/2
    mat17[i,300:350,1] = 255-i/2
    mat17[i,300:350,2] = 255-i/2

plt.imshow(mat17, cmap = "gray")
```
