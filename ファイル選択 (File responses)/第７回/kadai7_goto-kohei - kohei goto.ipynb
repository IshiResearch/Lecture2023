## 3-画像を四分割し，黄色、マゼンタ、シアン、灰色で表示する

```python
qua_img = np.zeros((10, 10, 3), np.uint8)
for row in range(10):
    for col in range(10):
        if ((row < 5) and (col < 5)): #左上->黄色
            qua_img[row, col, 0] = 255
            qua_img[row, col, 1] = 255
        elif ((row < 5) and (col >= 5)): #右上->マゼンタ
            qua_img[row, col, 0] = 255
            qua_img[row, col, 2] = 255
        elif ((row >= 5) and (col < 5)): #左下->シアン
            qua_img[row, col, 1] = 255
            qua_img[row, col, 2] = 255
        else: #右下->灰色
            qua_img[row, col, 0] = 128
            qua_img[row, col, 1] = 128
            qua_img[row, col, 2] = 128
            
plt.imshow(qua_img)
```

## 4-上から下に白色から黒色に変化する画像を作る

```python
grada_img = np.zeros((255, 255), np.uint8)
for row in range(255):
    grada_img[row, :] = 255 - row #255->白
    
plt.imshow(grada_img, "gray")
```

## 5-左から右に白から黒に変化する画像を作る

```python
slide_grada_img = np.zeros((255, 255), np.uint8)
for col in range(255):
    slide_grada_img[:, col] = 255 - col
    
plt.imshow(slide_grada_img, "gray")
```

## 6-左から右に黒から白に変化する画像を作る

```python
slide_rev_img = np.zeros((255, 255), np.uint8)
for col in range(255):
    slide_rev_img[:, col] = col
    
plt.imshow(slide_rev_img, "gray")
```

## 7-左上から右下に白から黒に変化する画像を作る

```python
diag_grada_img =  np.zeros((128, 128), np.uint8)
for row in range(128): #行
    for col in range(128): #列
        diag_grada_img[row, col] = 255 - (row + col)
        
plt.imshow(diag_grada_img, "gray")
```

## 8-左から右に黒から赤に変化する画像を作る

## 9-上から下に白から青に変化する画像を作る

```python
vert_blue_img = np.zeros((255, 255, 3), np.uint8)
vert_blue_img[:, :, :] = 255 #一度白い画像にする
for row in range(255):
    vert_blue_img[row, :, 0] = 255 - row #2(b)チャンネル以外を変化させる
    vert_blue_img[row, :, 1] = 255 - row
    
plt.imshow(vert_blue_img)
```

## 10-左から右に白からマゼンタに変化する画像を作る

## 11-縦に７分割して赤色，緑色、青色、黄色、マゼンタ、シアン、灰色が順に並ぶ画像を作る

```python
rainbow_img = np.zeros((255, 70, 3), np.uint8)
for color_num in range(7):
    for col in range(10):
        if (color_num == 0): #赤色
            rainbow_img[:, (col + (color_num * 10)), 0] = 255
        elif (color_num == 1): #緑色
            rainbow_img[:, (col + (color_num * 10)), 1] = 255
        elif (color_num == 2): #青色
            rainbow_img[:, (col + (color_num * 10)), 2] = 255
        elif (color_num == 3): #黄色
            rainbow_img[:, (col + (color_num * 10)), 0] = 255
            rainbow_img[:, (col + (color_num * 10)), 1] = 255
        elif (color_num == 4): #マゼンタ
            rainbow_img[:, (col + (color_num * 10)), 0] = 255
            rainbow_img[:, (col + (color_num * 10)), 2] = 255
        elif (color_num == 5): #シアン
            rainbow_img[:, (col + (color_num * 10)), 1] = 255
            rainbow_img[:, (col + (color_num * 10)), 2] = 255
        else: #灰色
            rainbow_img[:, (col + (color_num * 10)), :] = 128
            
plt.imshow(rainbow_img)
```

## 12-11番目の問題の画像を上から下に，白から徐々に色が変化していく画像を作る

```python
rainbow_grad_img = np.zeros((255, 70, 3), np.uint8)
rainbow_grad_img[:, :, :] = 255 #白い画像に変化させる
for color_num in range(7):
    for row in range(255):
        for col in range(10):
            if (color_num == 0):
                rainbow_grad_img[row, (col + (color_num * 10)), 0] = 255 - row
            elif (color_num == 1):
                rainbow_grad_img[row, (col + (color_num * 10)), 1] = 255 - row
            elif (color_num == 2):
                rainbow_grad_img[row, (col + (color_num * 10)), 2] = 255 - row
            elif (color_num == 3):
                rainbow_grad_img[row, (col + (color_num * 10)), 0] = 255 - row
                rainbow_grad_img[row, (col + (color_num * 10)), 1] = 255 - row
            elif (color_num == 4):
                rainbow_grad_img[row, (col + (color_num * 10)), 0] = 255 - row
                rainbow_grad_img[row, (col + (color_num * 10)), 2] = 255 - row
            elif (color_num == 5):
                rainbow_grad_img[row, (col + (color_num * 10)), 1] = 255 - row
                rainbow_grad_img[row, (col + (color_num * 10)), 2] = 255 - row
            else:
                rainbow_grad_img[row, (col + (color_num * 10)), :] = 255 - (row // 2) #128が灰色だから(//2)する
            
plt.imshow(rainbow_grad_img)
```

## 13-12で作成した画像を色ごとに表示する

```python
for i in range(7):
    div_img = np.zeros((255, 10, 3), np.uint8)
    for row in range(255):
        div_img[row, :] = rainbow_grad_img[row, (i * 10)]
    
    plt.imshow(div_img)
    plt.show() #一個ずつ表示するために必要、なかったら最後の画像だけ表示される
```
