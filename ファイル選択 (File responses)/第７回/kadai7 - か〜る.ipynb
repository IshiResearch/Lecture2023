### 11

```python
mat3d = np.zeros((256,280,3), np.uint8)

for i in range(3):
    mat3d[:, i*40:(i+1)*40, i] = 255
    
# 120~239列目:白　| 240~279列目：グレー
for j in range(3):
    mat3d[:, 120:240, j] = 255
    mat3d[:, 240:280, j] = 128
    
# 120~239列目　黄色、マゼンタ、シアンを作る
for k in range(3):
    mat3d[:, k*40+120:(k+1)*40+120, 2-k] = 0
    
plt.imshow(mat3d)
```

### 12

```python
def grad_color(a,x,y,z):
    for i in range(256):
        a[i, (x)*40:(x+1)*40, y] = 255 - i
        a[i, (x)*40:(x+1)*40, z] = 255 - i
        
def grad_color2(a,x,y):
    for i in range(256):
        a[i, (x)*40:(x+1)*40, y] = 255 - i
        
def grad_gray(a,v,x,y,z):
    for i in range(129):
        a[i, (v)*40:(v+1)*40, x] = 255 - i 
        a[i, (v)*40:(v+1)*40, y] = 255 - i
        a[i, (v)*40:(v+1)*40, z] = 255 - i

color_list = ["red", "green", "blue", "yellow", "magenta", "cyaan", "gray"]

grad_color(mat3d, color_list.index("red"), color_list.index("green"), color_list.index("blue"))
grad_color(mat3d, color_list.index("green"), color_list.index("red"), color_list.index("blue"))
grad_color(mat3d, color_list.index("blue"), color_list.index("red"), color_list.index("green"))
grad_color2(mat3d, color_list.index("yellow"), color_list.index("blue"))
grad_color2(mat3d, color_list.index("magenta"), color_list.index("green"))
grad_color2(mat3d, color_list.index("cyaan"), color_list.index("red"))
grad_gray(mat3d, color_list.index("gray"), color_list.index("red"), color_list.index("green"), color_list.index("blue"))

plt.imshow(mat3d)
```
