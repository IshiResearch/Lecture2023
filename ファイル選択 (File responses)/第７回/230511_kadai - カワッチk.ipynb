### 3. 画像を四分割し、黄色、マゼンタ、シアン、灰色で表示する

```python
mat_4cut = np.zeros((100, 100, 3), np.uint8)
mat_4cut[0:50, 50:100, 0] = 255
mat_4cut[50:100, 0:50, 0] = 255
mat_4cut[:, 0:50, 1]      = 255
mat_4cut[0:50, :, 2]      = 255
mat_4cut[50:100, 50:100]  = 180
plt.imshow(mat_4cut)
```

### 4. 上から下に白色から黒色に変化する画像を作る

```python
mat_grd = np.zeros((255, 255), np.uint8)
for i in range(255):
    mat_grd[i] = 255 - i
plt.imshow(mat_grd, "gray")
```

### 5. 左から右に白から黒に変化する画像を作る

```python
mat_grd2 = np.zeros((255, 255), np.uint8)
for i in range(255):
    for j in range(255):
        mat_grd2[i, j] = 255 - j
plt.imshow(mat_grd2, "gray")
```

### 6. 左から右に黒から白に変化する画像を作る

```python
mat_grd3 = np.zeros((255, 255), np.uint8)
for i in range(255):
    for j in range(255):
        mat_grd3[i, j] = j
plt.imshow(mat_grd3, "gray")
```

### 7. 左上から右下に白から黒に変化する画像を作る

```python
mat_grd4 = np.zeros((129, 128), np.uint8)
for i in range(129):
    white_num = 255 - i
    for j in range(128):
        mat_grd4[i, j] = white_num - j
plt.imshow(mat_grd4, "gray")
```

### 8. 左から右に黒から赤に変化する画像を作る

### 11. 縦に7分割して赤色、緑色、青色、黄色、マゼンタ、シアン、灰色が順に並ぶ画像を作る

```python
mat_rainbow = np.zeros((255, 350, 3), np.uint8)
mat_rainbow[:, 0:50, 0]    = 255
mat_rainbow[:, 150:250, 0] = 255
mat_rainbow[:, 50:100, 1]  = 255
mat_rainbow[:, 150:200, 1] = 255
mat_rainbow[:, 250:300, 1] = 255
mat_rainbow[:, 100:150, 2] = 255
mat_rainbow[:, 200:300, 2] = 255
mat_rainbow[:, 300:350]    = 160
plt.imshow(mat_rainbow)
```

### 12. 11番目の問題の画像を上から下に、白から徐々に色が変化していく画像を作る

```python
mat_rainbow2 = np.zeros((255, 350, 3), np.uint8)
for i in range(255):
    mat_rainbow2[i, 0:50, 0]    = 255
    mat_rainbow2[i, 50:150, 0]  = 255 - i
    mat_rainbow2[i, 150:250, 0] = 255
    mat_rainbow2[i, 250:300, 0] = 255 - i
    mat_rainbow2[i, 0:50, 1]    = 255 - i
    mat_rainbow2[i, 50:100, 1]  = 255
    mat_rainbow2[i, 100:150, 1] = 255 - i
    mat_rainbow2[i, 150:200, 1] = 255
    mat_rainbow2[i, 200:250, 1] = 255 - i
    mat_rainbow2[i, 250:300, 1] = 255
    mat_rainbow2[i, 0:100, 2]   = 255 - i
    mat_rainbow2[i, 100:150, 2] = 255
    mat_rainbow2[i, 150:200, 2] = 255 - i
    mat_rainbow2[i, 200:300, 2] = 255
    mat_rainbow2[i, 300:350]    = 255 - 0.5 * i
plt.imshow(mat_rainbow2)
```
