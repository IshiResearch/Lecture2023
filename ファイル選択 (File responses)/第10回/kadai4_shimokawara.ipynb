## 6.ノイズ除去

```python
img2 = np.random.randint(0,50,(501,501),np.uint8)
plt.imshow(img2,'gray')
rect_ker= np.ones((3,3))

ret,img2_thresh = cv2.threshold(img2,40,255, cv2.THRESH_BINARY)
plt.imshow(img2_thresh, 'gray')

img2_circle = cv2.circle(img2_thresh,(250,250),(150),(255,255,255),-1)
plt.imshow(img2_circle,'gray')

img2_erode = cv2.erode(img2_circle,rect_ker,iterations = 1)
plt.imshow(img2_erode,'gray')

img2_dilate = cv2.dilate(img2_erode,rect_ker,iterations = 1)
plt.imshow(img2_dilate,'gray')
```

## 7.car in showで物体検出

```python
img_car = "C:\\Users\\Shimo_Y_15\\20230519\\img\\carinsnow.jpg"
```

```python
img_bgr = cv2.imread(img_car)
imgrgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(imgrgb)
```

```python
img_car_gray = cv2.cvtColor(imgrgb,cv2.COLOR_RGB2GRAY)
plt.imshow(img_car_gray, 'gray')
```

```python
img_car_blur = cv2.blur(img_car_gray,(5,5))
plt.imshow(img_car_blur, 'gray')
```

```python
ret,img_car_thresh = cv2.threshold(img_car_blur,150,255,cv2.THRESH_BINARY_INV)
plt.imshow(img_car_thresh,'gray')
```

```python
img_car_erode = cv2.erode(img_car_thresh,(5,5),iterations = 5)
plt.imshow(img_car_erode, 'gray')
```

```python
img_car_dilate = cv2.dilate(img_car_erode,(5,5),iterations = 5)
plt.imshow(img_car_dilate, 'gray')
```

## 8.car in show2 を使い物体検出

```python
img_car2 = "C:\\Users\\Shimo_Y_15\\20230519\\img\\carinsnow2.jpg"
img_bgr2 = cv2.imread(img_car2)

imgrgb2 = cv2.cvtColor(img_bgr2,cv2.COLOR_BGR2RGB)

img_car2_gray = cv2.cvtColor(imgrgb2,cv2.COLOR_RGB2GRAY)

img_car2_blur = cv2.blur(img_car2_gray,(5,5))

ret,img_car2_thresh = cv2.threshold(img_car2_blur,140,255,cv2.THRESH_BINARY_INV)

img_car2_erode = cv2.erode(img_car2_thresh,(5,5),iterations = 5)

img_car2_dilate = cv2.dilate(img_car2_erode,(5,5),iterations = 5)

plt.imshow(img_car2_dilate,'gray')
```
