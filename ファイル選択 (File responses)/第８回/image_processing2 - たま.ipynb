```python
#3 好きな俳優の顔を切り取り、保存・表示する
img = cv2.imread("C:\\Users\\makot\\test\\img_2.jpg")
plt.imshow(img)
img_rgb = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img_rgb_clip = img_rgb[40:400,160:500]

plt.imshow(img_rgb_clip)
img_clip = cv2.cvtColor(img_rgb_clip, cv2.COLOR_BGR2RGB)
cv2.imwrite("C:\\Users\\makot\\test\\img_2_clip.jpg", img_clip)
```

```python
#4 いくつかの画像をダウンロードし、リサイズする（関数を用いて）
def Resize(path, save_path):
    img = cv2.imread(path)
    img_resize = cv2.resize(img, dsize=None, fx=0.25, fy=0.25)
    cv2.imwrite(save_path, img_resize)
    
filename = ["C:\\Users\\makot\\test\\img_g1.jpg", "C:\\Users\\makot\\test\\img_g2.jpg", "C:\\Users\\makot\\test\\img_g3.jpg"]
filename_save = ["C:\\Users\\makot\\test\\img_g1_resize.jpg", "C:\\Users\\makot\\test\\img_g2_resize.jpg", "C:\\Users\\makot\\test\\img_g3_resize.jpg"]

for i in range(len(filename)):
    Resize(filename[i], filename_save[i])
```

```python
#5 画像をフリップし、保存する
img_path = "C:\\Users\\makot\\test\\img_1.jpg"
img = cv2.imread(img_path)

for i in range(-1,2):
    img_flip = cv2.flip(img,i)
    save_path = "C:\\Users\\makot\\test\\flip"+str(i)+".jpg"
    cv2.imwrite(save_path, img_flip)
```

```python
#6 画像をダウンロードし、フリップして保存する
img = cv2.imread("C:\\Users\\makot\\test\\img_g1.jpg")
img_flip = cv2.flip(img, -1)
cv2.imwrite("C:\\Users\\makot\\test\\img_g1_flip.jpg", img_flip)
```

```python
#7 10％にして画像を保存・表示する
img = cv2.imread("C:\\Users\\makot\\test\\img_g1.jpg")
img_resize = cv2.resize(img, dsize=None, fx=0.1, fy=0.1)
cv2.imwrite("C:\\Users\\makot\\test\\img_g1_resize.jpg", img_resize)
img_resize_rgb = cv2.cvtColor(img_resize, cv2.COLOR_RGB2BGR)
plt.imshow(img_resize_rgb)
```
