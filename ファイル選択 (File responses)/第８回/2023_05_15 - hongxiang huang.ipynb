```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img = cv2.imread("C:\\img\\dog.jpg")
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img = np.array(img_rgb)

img_R = img.copy()
img_R[:, :, (1, 2)] = 0
img_G = img.copy()
img_G[:, :, (0, 2)] = 0
img_B = img.copy()
img_B[:, :, (0, 1)] = 0
plt.imshow(img_R)
cv2.imwrite("C:\\img\\img_r.jpg",img_R)
```

```python
huge = cv2.imread("C:\\img\\huge.jpg")
dog = cv2.imread("C:\\img\\dog.jpg")
jiraiya = cv2.imread("C:\\img\\jiraiya.jpg")


def Smaller(filename):
   
    for i in range(len(filename)):
        filename[i] = cv2.cvtColor(filename[i],cv2.COLOR_BGR2RGB)
        filename[i] = cv2.resize(filename[i],(0,0),filename[i],0.25,0.25)
      
        
filename = [huge,dog,jiraiya]
Smaller(filename)
plt.imshow(filename[0])
```

```python
def Gyaku(filename):
    for i in range(len(filename)):
        filename[i] = cv2.flip(filename[i],0)
Gyaku(filename) 
plt.imshow(filename[0])
 
```

```python
img = np.zeros(shape=(300,300,3),dtype=np.uint8)
plt.imshow(img)
```

```python
cv2.circle(img,(150,100),50,(255,0,0),-1)
plt.imshow(img)
```

```python
cv2.circle(img,(125,100),10,(255,255,255),-1)
cv2.circle(img,(175,100),10,(255,255,255),-1)
plt.imshow(img)
```

```python
cv2.rectangle(img,(100,150),(200,250),(255,0,0),-1)
plt.imshow(img)
```

```python
cv2.circle(img,(125,260),10,(255,0,0),-1)
cv2.circle(img,(175,260),10,(255,0,0),-1)
plt.imshow(img)
```

```python
img = cv2.line(img,(50,150),(100,200),(255,0,0),20)
img = cv2.line(img,(250,150),(200,200),(255,0,0),20)
plt.imshow(img)
```
