## 4-いくつかの画像をダウンロードし、ループを回してそれぞれのサイズを25％にする

```python
def resize(filename):
    resize_file = cv2.resize(filename, dsize=None, fx=0.25, fy=0.25)
    
    return(resize_file)
```

```python
path = "C:\\Users\\gokou\\practice\\0515_cv2\\img\\"
```

```python
img1 = cv2.imread(path + "dog.jpg")
img2 = cv2.imread(path + "lena.jpg")
img3 = cv2.imread(path + "parrot.jpg")

filename = [img1, img2, img3]

for i in range(3):
    resize_file = resize(filename[i])
    plt.imshow(resize_file)
    plt.show()
    
    cv2.imwrite(path + "resize\\img" + str(i) + ".jpg", resize_file)
```

## 8-ロボットの画像を作る

```python
robot_img = np.zeros((300, 300, 3), np.uint8)
cv2.circle(robot_img, (150, 100), (50), (255, 0, 0), -1)
plt.imshow(robot_img)
```

```python
cv2.rectangle(robot_img, (100, 150), (200, 250), (255, 0, 0), -1)
plt.imshow(robot_img)
```

```python
cv2.line(robot_img, (50, 150), (100, 200), (255, 0, 0), 20)
cv2.line(robot_img, (200, 200), (250, 150), (255, 0, 0), 20)
plt.imshow(robot_img)
```

```python
cv2.circle(robot_img, (120, 260), (10), (255, 0, 0), -1)
cv2.circle(robot_img, (180, 260), (10), (255, 0, 0), -1)
cv2.circle(robot_img, (130, 100), (10), (255, 255, 255), -1)
cv2.circle(robot_img, (170, 100), (10), (255, 255, 255), -1)
plt.imshow(robot_img)
```

```python
cv2.putText(robot_img, "Robot", (50, 250), cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), 3, cv2.LINE_AA)
plt.imshow(robot_img)
```
