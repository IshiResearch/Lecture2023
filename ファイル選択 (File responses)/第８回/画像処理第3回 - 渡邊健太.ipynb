```python
#thresholdthresh0,50,128,200vconcat
img_path = "C:\\Users\\smart27\\Dropbox\\python\\img (2)\\dog193.jpg"
dog_img = cv2.imread(img_path)
dog_rgb = cv2.cvtColor(dog_img, cv2.COLOR_BGR2RGB)
plt.imshow(dog_rgb)
```

```python
dog_gray = cv2.cvtColor(dog_rgb,cv2.COLOR_BGR2GRAY)
plt.imshow(dog_gray,cmap="gray")
```

```python
ret, dog_thresh0 = cv2.threshold(dog_gray, 0, 255, cv2.THRESH_BINARY) #(, threshold(), maxvalue, thresholdtype)

ret, dog_thresh55 = cv2.threshold(dog_gray, 55, 255, cv2.THRESH_BINARY)

ret, dog_thresh128 = cv2.threshold(dog_gray, 128, 255, cv2.THRESH_BINARY)

ret, dog_thresh200 = cv2.threshold(dog_gray, 200, 255, cv2.THRESH_BINARY)

img_concat = cv2.vconcat([dog_thresh0, dog_thresh55, dog_thresh128, dog_thresh200])
plt.figure(figsize = (6,12))
plt.axis("off")
plt.imshow(img_concat, cmap = "gray")
```

```python
#thresholdTypeTHRESH_OTSU,THRESH_TRUNK,THRESH_TOZERO,THRESH_TOZERO_INV
img_path = "C:\\Users\\smart27\\Dropbox\\python\\img (2)\\dog193.jpg"
dog_img = cv2.imread(img_path)
dog_rgb = cv2.cvtColor(dog_img, cv2.COLOR_BGR2RGB)

dog_gray = cv2.cvtColor(dog_rgb,cv2.COLOR_BGR2GRAY)
```

```python
ret, dog_thresh_otsu = cv2.threshold(dog_gray, 200, 255, cv2.THRESH_OTSU) 

ret, dog_thresh_tozero = cv2.threshold(dog_gray, 200, 255, cv2.THRESH_TOZERO) 

img_concat_l = cv2.vconcat([dog_thresh_otsu, dog_thresh_tozero])
```

```python
ret, dog_thresh_trunc = cv2.threshold(dog_gray, 200, 255, cv2.THRESH_TRUNC) 

ret, dog_thresh_tozero_inv = cv2.threshold(dog_gray, 200, 255, cv2.THRESH_TOZERO_INV) 

img_concat_r = cv2.vconcat([dog_thresh_trunc, dog_thresh_tozero_inv])
```

```python
img_concat_all = cv2.hconcat([img_concat_l, img_concat_r])
plt.figure(figsize = (12,6))
plt.axis("off")
plt.imshow(img_concat_all, cmap= "gray")
```

```python
#threshold,gray,HSV,BGR
img_path = "C:\\Users\\smart27\\Dropbox\\python\\img (2)\\dog193.jpg"
dog_img = cv2.imread(img_path)
dog_rgb = cv2.cvtColor(dog_img, cv2.COLOR_BGR2RGB)

dog_gray = cv2.cvtColor(dog_rgb,cv2.COLOR_BGR2GRAY)
dog_gray_edge = cv2.Canny(dog_gray, 80, 200)
#plt.imshow(dog_gray_edge)
dog_gray_3ch = np.stack((dog_gray_edge,)*3, -1)
```

```python
ret, dog_thresh = cv2.threshold(dog_gray, 200, 255, cv2.THRESH_BINARY)
dog_thresh_c = cv2.cvtColor(dog_thresh, cv2.COLOR_GRAY2RGB)
print(dog_thresh_c.shape)

dog_hsv = cv2.cvtColor(dog_rgb, cv2.COLOR_RGB2HSV)

base = np.zeros(shape=(360,287,3), dtype=np.uint8)
base[:base.shape[0]//2, :base.shape[1]//2] = dog_rgb[:dog_rgb.shape[0]//2, :dog_gray.shape[1]//2] #RGB()

base[:base.shape[0]//2, base.shape[1]//2:] = dog_thresh_c[:dog_thresh_c.shape[0]//2, dog_thresh_c.shape[1]//2:] #thresh()

base[base.shape[0]//2:, :base.shape[1]//2] = dog_gray_3ch[dog_gray_3ch.shape[0]//2:, :dog_gray_3ch.shape[1]//2] #gray()

base[base.shape[0]//2:, base.shape[1]//2:] = dog_hsv[dog_hsv.shape[0]//2:, dog_hsv.shape[1]//2:] #HSV()
plt.imshow(base, cmap="gray")
```

```python
base_blend = cv2.addWeighted(dog_img, 0.5, base, 0.5, gamma=0)
plt.imshow(base_blend)
```
