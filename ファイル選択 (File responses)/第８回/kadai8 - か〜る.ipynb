### 4

```python
animal_list = ["dog", "panda", "rabbit", "squirre"]
filename = []

for i in animal_list:
    img_path = "C:\\img\\{}.jpg".format(i)
    img = cv2.imread(img_path)
    resize_i = cv2.resize(img, dsize=None, fx=0.25,fy=0.25)
    filename.append(resize_i)
```

### 5

```python
for i in range(len(filename)):
    img_flip = cv2.flip(filename[i],0)
    save_path = "C:\\img\\filp{}.jpg".format(i)
    cv2.imwrite(save_path,img_flip)
```
