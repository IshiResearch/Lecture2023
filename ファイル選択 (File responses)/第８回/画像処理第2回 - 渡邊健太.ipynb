```python
dir_name = "C:\\Users\\smart27\\Dropbox\\python\\img (2)\\img_4"
img1 = dir_name + "\\img1.jpg"
img2 = dir_name + "\\img2.jpg"
img3 = dir_name + "\\img3.jpg"
img4 = dir_name + "\\img4.jpg"
img5 = dir_name + "\\img5.jpg"
img6 = dir_name + "\\img6.jpg"
filenames = [img1,img2,img3,img4,img5,img6] 
img = ""
def quarter(filenames):
    for file_name in filenames:
        img = cv2.imread(file_name)
        img_re = cv2.resize(img,None,None,0.2,0.2)
        write_name = file_name.split(".")[0] + "_resize." + file_name.split(".")[1]
        cv2.imwrite(write_name,img_re)
    return filenames
    
        
    
     
```
