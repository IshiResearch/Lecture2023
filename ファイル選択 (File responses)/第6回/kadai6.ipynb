```python
class Complex():
    def __init__(self, real_num, imaginaly_num): # 初期値の設定
        self.real_num = real_num
        self.imaginaly_num = imaginaly_num
        
    def __add__(self,other):
        return self.real_num + other.real_num, self.imaginaly_num + other.imaginaly_num
    
    def __sub__(self,other):
        return self.real_num - other.real_num, self.imaginaly_num - other.imaginaly_num
        
    def __mul__(self,other):
        return self.real_num*other.real_num - self.imaginaly_num*other.imaginaly_num, self.real_num*other.imaginaly_num + self.imaginaly_num*other.real_num
         
    def __truediv__(self,other):
        return (self.real_num*other.real_num + self.imaginaly_num*other.imaginaly_num)/((self.real_num**2)+(self.imaginaly_num**2)), (self.real_num*other.imaginaly_num - self.imaginaly_num*other.real_num)/((self.real_num**2)+(self.imaginaly_num**2))

a_1,b_1 = 3,2
a_2,b_2 = 2,1

z_1 = Complex(a_1, b_1)
z_2 = Complex(a_2, b_2)

z_add     = z_1 + z_2
z_sub     = z_1 - z_2
z_mul     = z_1 * z_2
z_truediv = z_1 / z_2

print(f"add:{z_add[0]} + {z_add[1]}i")
print(f"sub:{z_sub[0]} + {z_sub[1]}i")
print(f"mul:{z_mul[0]} + {z_mul[1]}i")
print(f"div:{z_truediv[0]} + {z_truediv[1]}i")
```

```python
class Car():
    def __init__(self, distance, color, speed, fuel, fue_con, refuel):
        self.distance = distance   # 距離
        self.color    = color      # 色
        self.speed    = speed      # 速さ
        self.fuel     = fuel       # 燃料の残量
        self.fue_con  = fue_con    # 燃費
        self.refuel   = refuel     # 給油速度
        self.max_fuel = fuel       # maxの燃料
        self.switch   = 0
    
    def accelerate(self): # 走らせる
        self.distance -= self.speed
        
    def fuel_amount(self): # 燃料を減らす
        self.fuel -= self.fue_con
    
    def refueling(self): # 給油
        self.fuel += self.refuel
        
    def switching(self): # 燃料が0になったらself.switchを1に,満タンになったら0にする
        if self.fuel <= 0:
            self.switch += 1
        
        elif (self.switch == 1) and (self.fuel == self.max_fuel):
            self.switch -= 1
    
    def info(self):
        print(f"{self.color} : ゴールまで{self.distance}m / 残り{self.fuel}ml")
    
    def goal(self):
        print(f"{self.color} Goal !!")
    
    def step(self):
        self.switching()
        if self.switch == 1: # switchが1の間は給油
            self.refueling()
            self.info()
            
        else:
            self.accelerate()
            self.fuel_amount()
            self.info()
            
        if self.distance <= 0:
            self.goal()

    
def main(r,b,g):
    r.step()
    b.step()
    g.step()
    
distance = 1000

red   = Car(distance, "Red", 30, 1000, 100, 200)
blue  = Car(distance, "Blue", 40, 500, 100, 100)
green = Car(distance, "Green", 50, 300, 100, 100)

while (red.distance > 0) and (blue.distance > 0) and (green.distance > 0):
    main(red, blue, green)
            
```

```python
class Car():
    def __init__(self, distance, color, speed, fuel, fue_con, refuel):
        self.distance = distance   # 距離
        self.color = color         # 色
        self.speed = speed         # 速さ
        self.fuel = fuel           # 燃料の残量
        self.fue_con = fue_con     # 燃費
        self.refuel = refuel # 給油速度
        self.max_fuel = fuel # maxの燃料
        self.switch = 0
        self.letter = ""
    
    def accelerate(self): # 走らせる
        self.distance -= self.speed
        
    def fuel_amount(self): # 燃料を減らす
        self.fuel -= self.fue_con
    
    def refueling(self): # 給油
        self.fuel += self.refuel
        
    def switching(self): # 燃料が0になったらself.switchを1に,満タンになったら0にする
        if self.fuel <= 0:
            self.switch += 1
        
        elif (self.switch == 1) and (self.fuel == self.max_fuel):
            self.switch -= 1
    
    def info(self): # 情報の提示
        print(f"{self.color} : ゴールまで{self.distance}m | 残り{self.fuel}ml")
    
    def goal(self): # ゴールの出力
        print(f"{self.color} Goal !!")
        
            
    def keep_on_refueling(self):
        if self.switch == 1: # switchが1の間は給油
            self.refueling()
            self.info()
            
    def print_car_info(self): # 車の情報の出力
        if self.letter == "f":
            print(f"<{self.color}>\n速さ : {self.speed}m/step\n燃料の残量 : {self.fuel}ml\n燃費 : {self.fue_con}ml/step")
            
    def error_refueling(self): # 燃料が満タンの時に給油した場合のエラー
        if (self.fuel == self.max_fuel) and (self.letter == "s"):
            print("これ以上給油できません。")
            while self.letter != "w":
                self.letter = input(f"{self.color}: w = アクセル")
            
    def input_letter(self): # 入力によってアクセルか給油を選択
        self.letter = input(f"{self.color}: w = アクセル | s = 給油 | f = 情報")
        
        if self.letter == "f":
            self.print_car_info()
            self.letter = input(f"{self.color}: w = アクセル | s = 給油")
        
        while (self.letter != "w") and (self.letter != "s"):
            print("入力が不正です")
            self.letter = input(f"{self.color}: w = アクセル | s = 給油")
    
    def main_step(self):
        self.switching()
        self.keep_on_refueling()
        self.input_letter()
        self.error_refueling()
        
        if self.letter == "w":
            self.accelerate()
            self.fuel_amount()
            self.info()
        
        else:
            self.refueling()
            self.info()
            
        if self.distance <= 0:
            self.goal()

    
def main(r,b,g):
    r.main_step()
    b.main_step()
    g.main_step()
    
distance = 30

red = Car(distance, "Red", 30, 1000, 100, 200)
blue = Car(distance, "Blue", 40, 500, 100, 100)
green = Car(distance, "Green", 50, 300, 100, 100)

while (red.distance > 0) and (blue.distance > 0) and (green.distance > 0):
    main(red, blue, green)
            
```
