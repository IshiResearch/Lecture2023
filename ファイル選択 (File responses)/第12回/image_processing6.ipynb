```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 fps=1, 0.2, 10でカメラから画像を表示
cap = cv2.VideoCapture(0)

while cap.isOpened():
    retval, frame = cap.read()
    if retval is None:
        print("capture error")
        break

    cv2.imshow("capture", frame)

    fps = 1000//1

    key = cv2.waitKey(int(fps))
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#2 1fpsでカメラ画像、グレー画像をそれぞれ保存
cap = cv2.VideoCapture(0)
num = 0

while cap.isOpened():
    retval, frame = cap.read()
    if retval is None:
        print("capture error")
        break
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    num += 1

    cv2.imshow("capture", frame)
    cv2.imshow("capture_gray", frame_gray)

    cv2.imwrite("C:\\Users\\makot\\test\\frame\\"+str(num)+".jpg", frame)
    cv2.imwrite("C:\\Users\\makot\\test\\frame_gray\\"+str(num)+".jpg", frame_gray)

    fps = 1000//1

    key = cv2.waitKey(int(fps))
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#3 1fpsで平滑化画像、グレー画像、二値化画像を表示
import cv2
cap = cv2.VideoCapture(0)

while cap.isOpened():
    retval, frame = cap.read()
    if retval is None:
        print("capture error")
        break
    frame_blur = cv2.blur(frame, (11,11))
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    ret, frame_thresh = cv2.threshold(frame_gray, 127, 255, cv2.THRESH_BINARY_INV)
    
    cv2.imshow("capture_blur", frame_blur)
    cv2.imshow("capture_gray", frame_gray)
    cv2.imshow("capture_thresh", frame_thresh)

    fps = 1000//1

    key = cv2.waitKey(int(fps))
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#4 1fpsで好きなものを検出した画像を表示
cap = cv2.VideoCapture(0)

img = cv2.imread("C:\\Users\\makot\\img\\WIN_20230526_10_33_04_Pro.jpg")
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
H,S,V = img_hsv[450, 1000]
upper = np.array([H+5, S+30, V+50])
lower = np.array([H-5, S-30, V-50])

while cap.isOpened():
    retval, frame = cap.read()
    if retval is None:
        print("capture error")
        break

    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    frame_blur = cv2.blur(frame_hsv, (11,11))
    frame_range = cv2.inRange(frame_blur, lower, upper)

    contours, hierarchy = cv2.findContours(frame_range, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for each_contours in contours:
        x,y,w,h = cv2.boundingRect(each_contours)
        if w*h >= 1000:
            frame_rect = cv2.rectangle(frame, (x,y), (x+w,y+h), (255,0,0), 2)
    
    cv2.imshow("capture", frame)

    key = cv2.waitKey(100)
    if key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#5 10fpsでカメラから手のひらを検出した画像を表示
cap = cv2.VideoCapture(0)

img = cv2.imread("C:\\Users\\makot\\img\\WIN_20230526_11_10_43_Pro.jpg")
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hsv_blur = cv2.blur(img_hsv, (19,19))
H,S,V = img_hsv_blur[442,322]
lower = np.array([H-5, S-30, V-50])
upper = np.array([H+5, S+30, V+50])

while cap.isOpened():
    retval, frame = cap.read()
    if retval is None:
        print("capture error")
        break
    frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    frame_hsv_blur = cv2.blur(frame_hsv, (19,19))
    
    frame_range = cv2.inRange(frame_hsv_blur, lower, upper)

    frame_erode = cv2.erode(frame_range, np.ones((5,5), np.uint8), iterations=4)
    frame_dilate = cv2.dilate(frame_erode, np.ones((5,5), np.uint8), iterations=4)

    contours, hierarchy = cv2.findContours(frame_dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    for each_contours in contours:
        x,y,w,h = cv2.boundingRect(each_contours)
        if w*h >= 10000:
            frame_rect = cv2.rectangle(frame, (x,y), (x+w,y+h), (0,0,255), 2)
            
    cv2.imshow("capture_range", frame_range)
    cv2.imshow("capture_erode", frame_erode)
    cv2.imshow("capture_dilate", frame_dilate)

    cv2.imshow("capture", frame)
    
    key = cv2.waitKey(100)
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#6 10fpsでカメラからflipした画像をそれぞれ表示
cap = cv2.VideoCapture(0)

while cap.isOpened():
    retval, frame = cap.read()
    if retval is None:
        print("capture error")
        break
    frame_flip_1 = cv2.flip(frame, 1)
    frame_flip_0 = cv2.flip(frame, 0)
    frame_flip_minus1 = cv2.flip(frame, -1)
    
    cv2.imshow("capture_origin", frame)
    cv2.imshow("capture_flip_1", frame_flip_1)
    cv2.imshow("capture_flip_0", frame_flip_0)
    cv2.imshow("capture_flip_minus1", frame_flip_minus1)

    key = cv2.waitKey(100)
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
