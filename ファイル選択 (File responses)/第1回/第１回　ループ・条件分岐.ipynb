```python
#2
japanese = float(input("国語"))
math = float(input("数学"))
physics = float(input("物理"))
science = float(input("化学"))
english = float(input("英語"))

list = [japanese,math,physics,science,english]
ave = sum(list) / len(list)
print(ave)

if ave >= 90:
    print("秀")
elif 80 < ave <= 89:
    print("優")
elif 70 < ave <= 79:
    print("良")
elif 60 < ave <= 69:
    print("可")
elif ave < 60:
    print("不可")
else :
    print("測定不可")
```

```python
#5
r = float(input("半径")) 
x = float(input("中心x"))
y = float(input("中心y"))
x1 = float(input("x座標を入力"))
y1 = float(input("y座標を入力"))
if (x1-x)**2.0 + (y1-y)**2.0 > r**2.0:
    print("円の外")
    

elif (x1-x)**2.0 + (y1-y)**2.0 < r**2.0:
    print("円の中")
    
else:
    print("円の上")
```
